## 3ï¸âƒ£ Veri YapÄ±larÄ± - Liste (List) - DetaylÄ±

**Liste**, birden fazla veriyi sÄ±ralÄ± bir ÅŸekilde saklamamÄ±zÄ± saÄŸlar. KÃ¶ÅŸeli parantez `[]` kullanÄ±lÄ±r. Listeler **deÄŸiÅŸtirilebilir (mutable)** yapÄ±lardÄ±r.

### Liste OluÅŸturma YÃ¶ntemleri
```python
liste1 = [1, 2, 3, 4, 5]  # DoÄŸrudan
liste2 = list(range(1, 6))  # range() kullanarak
liste3 = []  # BoÅŸ liste
```

### Temel Liste Ä°ÅŸlemleri
- **Ekleme**: `liste.append(6)` - Listenin sonuna ekler
- **EriÅŸim**: `liste[0]` - Ä°lk elemana eriÅŸir (indeks 0'dan baÅŸlar!)
- **Uzunluk**: `len(liste)` - Listenin kaÃ§ elemanÄ± olduÄŸunu gÃ¶sterir
- **Silme**: `liste.remove(3)` - Belirli bir deÄŸeri siler
- **Ä°ndeks ile Silme**: `del liste[0]` - Belirli bir indeksteki elemanÄ± siler
- **SÄ±ralama**: `liste.sort()` - Listeyi sÄ±ralar
- **Ters Ã‡evirme**: `liste.reverse()` - Listeyi ters Ã§evirir


````python
# Liste Ã¶rnekleri - KapsamlÄ±

print("=" * 60)
print("LÄ°STE Ä°ÅLEMLERÄ° - DETAYLI Ã–RNEKLER")
print("=" * 60)
print()

# 1. Liste oluÅŸturma yÃ¶ntemleri
print("1. LÄ°STE OLUÅTURMA YÃ–NTEMLERÄ°")
print("-" * 40)
liste1 = [1, 2, 3, 4, 5]
liste2 = list(range(1, 6))
liste3 = []
print(f"DoÄŸrudan: {liste1}")
print(f"range() ile: {liste2}")
print(f"BoÅŸ liste: {liste3}")
print()

# 2. Liste oluÅŸturma
isimler = ["Ali", "AyÅŸe", "Mehmet", "Zeynep", "Can"]
print("2. LÄ°STE ELEMANLARINA ERÄ°ÅÄ°M")
print("-" * 40)
print(f"TÃ¼m liste: {isimler}")
print(f"Ä°lk Ã¶ÄŸrenci (indeks 0): {isimler[0]}")
print(f"Ä°kinci Ã¶ÄŸrenci (indeks 1): {isimler[1]}")
print(f"Son Ã¶ÄŸrenci (indeks -1): {isimler[-1]}")
print(f"Ä°kinci son (indeks -2): {isimler[-2]}")
print(f"Ä°lk 3 Ã¶ÄŸrenci (slicing): {isimler[0:3]}")
print(f"Son 2 Ã¶ÄŸrenci: {isimler[-2:]}")
print()

# 3. Listeye eleman ekleme
print("3. LÄ°STEYE ELEMAN EKLEME")
print("-" * 40)
print(f"BaÅŸlangÄ±Ã§: {isimler}")
isimler.append("Elif")  # Sonuna ekle
print(f"append() sonrasÄ±: {isimler}")
isimler.insert(0, "Zeynep")  # Belirli bir pozisyona ekle
print(f"insert(0, 'Zeynep') sonrasÄ±: {isimler}")
isimler.extend(["Fatma", "Hasan"])  # Birden fazla eleman ekle
print(f"extend() sonrasÄ±: {isimler}")
print()

# 4. Listeden eleman silme
print("4. LÄ°STEDEN ELEMAN SÄ°LME")
print("-" * 40)
print(f"BaÅŸlangÄ±Ã§: {isimler}")
isimler.remove("Zeynep")  # Ä°lk bulduÄŸu "Zeynep"i siler
print(f"remove('Zeynep') sonrasÄ±: {isimler}")
silinen = isimler.pop()  # Son elemanÄ± siler ve dÃ¶ndÃ¼rÃ¼r
print(f"pop() sonrasÄ± (silinen: {silinen}): {isimler}")
del isimler[0]  # Ä°ndeks ile silme
print(f"del isimler[0] sonrasÄ±: {isimler}")
print()

# 5. Liste sÄ±ralama ve arama
print("5. LÄ°STE SIRALAMA VE ARAMA")
print("-" * 40)
sayilar = [5, 2, 8, 1, 9, 3]
print(f"BaÅŸlangÄ±Ã§: {sayilar}")
sayilar.sort()  # KÃ¼Ã§Ã¼kten bÃ¼yÃ¼ÄŸe sÄ±rala
print(f"sort() sonrasÄ±: {sayilar}")
sayilar.sort(reverse=True)  # BÃ¼yÃ¼kten kÃ¼Ã§Ã¼ÄŸe sÄ±rala
print(f"sort(reverse=True) sonrasÄ±: {sayilar}")
print(f"'AyÅŸe' listede var mÄ±? {'AyÅŸe' in isimler}")
print(f"'AyÅŸe' indeksi: {isimler.index('AyÅŸe') if 'AyÅŸe' in isimler else 'BulunamadÄ±'}")
print()

# 6. Liste uzunluÄŸu ve diÄŸer iÅŸlemler
print("6. DÄ°ÄER Ä°ÅLEMLER")
print("-" * 40)
print(f"Toplam Ã¶ÄŸrenci sayÄ±sÄ±: {len(isimler)}")
print(f"Liste boÅŸ mu? {len(isimler) == 0}")
print(f"En uzun isim: {max(isimler, key=len)}")
print(f"En kÄ±sa isim: {min(isimler, key=len)}")
````

## 4ï¸âƒ£ Veri YapÄ±larÄ± - Tuple (Demet) - YENÄ°!

**Tuple**, liste gibi ama **deÄŸiÅŸtirilemez (immutable)** bir veri yapÄ±sÄ±dÄ±r. Parantez `()` kullanÄ±lÄ±r.

### Tuple KullanÄ±m AlanlarÄ±:
- Koordinatlar: `(x, y)`
- RGB renkler: `(255, 0, 0)`
- DeÄŸiÅŸmemesi gereken veriler
- Daha hÄ±zlÄ± eriÅŸim gerektiÄŸinde

### Tuple Ä°ÅŸlemleri:
- **OluÅŸturma**: `tuple = (1, 2, 3)`
- **EriÅŸim**: `tuple[0]` - Liste gibi eriÅŸim
- **Unpacking**: `x, y = (10, 20)` - DeÄŸerleri ayrÄ± deÄŸiÅŸkenlere atama


````python
# Tuple Ã¶rnekleri

print("=" * 60)
print("TUPLE (DEMET) Ä°ÅLEMLERÄ°")
print("=" * 60)
print()

# 1. Tuple oluÅŸturma
print("1. TUPLE OLUÅTURMA")
print("-" * 40)
koordinat = (10, 20)
renk_rgb = (255, 0, 0)  # KÄ±rmÄ±zÄ±
print(f"Koordinat: {koordinat}")
print(f"RGB Renk: {renk_rgb}")
print()

# 2. Tuple eriÅŸim
print("2. TUPLE ERÄ°ÅÄ°M")
print("-" * 40)
print(f"X koordinatÄ±: {koordinat[0]}")
print(f"Y koordinatÄ±: {koordinat[1]}")
print(f"Son eleman: {koordinat[-1]}")
print()

# 3. Tuple unpacking (Ã§ok Ã¶nemli!)
print("3. TUPLE UNPACKING")
print("-" * 40)
x, y = koordinat
print(f"x = {x}, y = {y}")

r, g, b = renk_rgb
print(f"R={r}, G={g}, B={b}")
print()

# 4. Tuple vs List - Fark
print("4. TUPLE vs LÄ°STE FARKI")
print("-" * 40)
liste = [1, 2, 3]
demet = (1, 2, 3)
print(f"Liste: {liste}, DeÄŸiÅŸtirilebilir")
liste[0] = 99  # DeÄŸiÅŸtirilebilir
print(f"Liste deÄŸiÅŸtirildi: {liste}")

print(f"Demet: {demet}, DeÄŸiÅŸtirilemez")
# demet[0] = 99  # HATA! Tuple deÄŸiÅŸtirilemez
print("Tuple deÄŸiÅŸtirilemez - yorum satÄ±rÄ±nÄ± kaldÄ±rÄ±rsanÄ±z hata alÄ±rsÄ±nÄ±z")
print()

# 5. Tuple kullanÄ±m Ã¶rnekleri
print("5. PRATÄ°K KULLANIM Ã–RNEKLERÄ°")
print("-" * 40)
ogrenci_bilgisi = ("Ahmet", 20, 85.5)
isim, yas, notu = ogrenci_bilgisi
print(f"Ã–ÄŸrenci: {isim}, YaÅŸ: {yas}, Not: {notu}")
````

## 5ï¸âƒ£ Veri YapÄ±larÄ± - Set (KÃ¼me) - YENÄ°!

**Set**, **tekrarsÄ±z** elemanlar iÃ§eren bir veri yapÄ±sÄ±dÄ±r. SÃ¼slÃ¼ parantez `{}` kullanÄ±lÄ±r.

### Set KullanÄ±m AlanlarÄ±:
- TekrarsÄ±z veri saklama
- Matematiksel kÃ¼me iÅŸlemleri (birleÅŸim, kesiÅŸim, fark)
- HÄ±zlÄ± Ã¼yelik kontrolÃ¼

### Set Ä°ÅŸlemleri:
- **OluÅŸturma**: `kume = {1, 2, 3}`
- **Ekleme**: `kume.add(4)`
- **Silme**: `kume.remove(2)`
- **KesiÅŸim**: `kume1 & kume2`
- **BirleÅŸim**: `kume1 | kume2`
- **Fark**: `kume1 - kume2`


````python
# Set Ã¶rnekleri

print("=" * 60)
print("SET (KÃœME) Ä°ÅLEMLERÄ°")
print("=" * 60)
print()

# 1. Set oluÅŸturma (tekrarlar otomatik silinir)
print("1. SET OLUÅTURMA")
print("-" * 40)
sayilar = {1, 2, 3, 4, 5, 3, 2, 1}  # Tekrarlar silinir
print(f"SayÄ±lar (tekrarlar silindi): {sayilar}")
isimler_set = {"Ali", "AyÅŸe", "Mehmet", "Ali"}  # "Ali" tekrarÄ± silindi
print(f"Ä°simler: {isimler_set}")
print()

# 2. Set iÅŸlemleri
print("2. SET Ä°ÅLEMLERÄ°")
print("-" * 40)
sayilar.add(6)
print(f"6 eklendikten sonra: {sayilar}")
sayilar.remove(1)
print(f"1 silindikten sonra: {sayilar}")
print(f"Set uzunluÄŸu: {len(sayilar)}")
print(f"5 set'te var mÄ±? {5 in sayilar}")
print()

# 3. Matematiksel kÃ¼me iÅŸlemleri
print("3. MATEMATÄ°KSEL KÃœME Ä°ÅLEMLERÄ°")
print("-" * 40)
kume1 = {1, 2, 3, 4, 5}
kume2 = {4, 5, 6, 7, 8}

print(f"KÃ¼me 1: {kume1}")
print(f"KÃ¼me 2: {kume2}")
print(f"KesiÅŸim (ortak elemanlar): {kume1 & kume2}")
print(f"BirleÅŸim (tÃ¼m elemanlar): {kume1 | kume2}")
print(f"Fark (kume1'de olup kume2'de olmayan): {kume1 - kume2}")
print(f"Simetrik fark (her ikisinde de olmayan): {kume1 ^ kume2}")
print()

# 4. Pratik kullanÄ±m - TekrarsÄ±z veri
print("4. PRATÄ°K KULLANIM - TEKRARSIZ VERÄ°")
print("-" * 40)
notlar_listesi = [85, 90, 78, 85, 92, 90, 78]
print(f"Orijinal liste (tekrarlÄ±): {notlar_listesi}")
benzersiz_notlar = set(notlar_listesi)
print(f"Benzersiz notlar: {benzersiz_notlar}")
print(f"Benzersiz not sayÄ±sÄ±: {len(benzersiz_notlar)}")
````

## 6ï¸âƒ£ Veri YapÄ±larÄ± - SÃ¶zlÃ¼k (Dictionary) - DetaylÄ±

**SÃ¶zlÃ¼k**, verileri **anahtar-deÄŸer** Ã§iftleri halinde saklar. GerÃ§ek bir sÃ¶zlÃ¼k gibi dÃ¼ÅŸÃ¼nebilirsiniz:
- **Anahtar (Key)**: Kelime (Ã¶rn: "isim")
- **DeÄŸer (Value)**: AÃ§Ä±klama (Ã¶rn: "Ahmet")

### SÃ¶zlÃ¼k OluÅŸturma YÃ¶ntemleri:
```python

```

### SÃ¶zlÃ¼k Ä°ÅŸlemleri:
- **EriÅŸim**: `sozluk["anahtar"]` veya `sozluk.get("anahtar")`
- **Ekleme/GÃ¼ncelleme**: `sozluk["yeni_anahtar"] = "yeni_deÄŸer"`
- **Silme**: `del sozluk["anahtar"]` veya `sozluk.pop("anahtar")`
- **TÃ¼m anahtarlar**: `sozluk.keys()`

## 1ï¸âƒ£7ï¸âƒ£ Pandas GroupBy - Gruplama ve Toplama

GroupBy, verileri gruplara ayÄ±rÄ±p her grup Ã¼zerinde iÅŸlem yapmamÄ±zÄ± saÄŸlar. Excel'deki pivot tablolara benzer!


````python
# Pandas GroupBy Ä°ÅŸlemleri

print("=" * 60)
print("PANDAS GROUPBY - GRUPLAMA VE TOPLAMA")
print("=" * 60)
print()

# Ã–rnek veri
df = pd.DataFrame({
    'BÃ¶lÃ¼m': ['Bilgisayar', 'Elektrik', 'Bilgisayar', 'EndÃ¼stri', 'Makine', 
              'Bilgisayar', 'Elektrik', 'EndÃ¼stri'],
    'Ä°sim': ['Ahmet', 'AyÅŸe', 'Mehmet', 'Zeynep', 'Can', 'Elif', 'Hasan', 'Fatma'],
    'Not': [85, 92, 78, 95, 88, 90, 82, 87],
    'YaÅŸ': [20, 21, 19, 20, 22, 19, 21, 20]
})

print("Orijinal DataFrame:")
print(df)
print()

# 1. Basit gruplama
print("1. BASÄ°T GRUPLAMA")
print("-" * 40)
bolum_grup = df.groupby('BÃ¶lÃ¼m')
print("BÃ¶lÃ¼mlere gÃ¶re gruplandÄ±:")
for isim, grup in bolum_grup:
    print(f"\n{isim} BÃ¶lÃ¼mÃ¼:")
    print(grup)
print()

# 2. Gruplama ve toplama
print("2. GRUPLAMA VE TOPLAMA")
print("-" * 40)
print("BÃ¶lÃ¼mlere gÃ¶re ortalama not:")
print(df.groupby('BÃ¶lÃ¼m')['Not'].mean())
print("\nBÃ¶lÃ¼mlere gÃ¶re Ã¶ÄŸrenci sayÄ±sÄ±:")
print(df.groupby('BÃ¶lÃ¼m').size())
print("\nBÃ¶lÃ¼mlere gÃ¶re tÃ¼m istatistikler:")
print(df.groupby('BÃ¶lÃ¼m')['Not'].describe())
print()

# 3. Ã‡oklu toplama fonksiyonlarÄ±
print("3. Ã‡OKLU TOPLAMA FONKSÄ°YONLARI")
print("-" * 40)
toplam = df.groupby('BÃ¶lÃ¼m')['Not'].agg(['mean', 'min', 'max', 'count'])
print("BÃ¶lÃ¼mlere gÃ¶re not istatistikleri:")
print(toplam)
print()

# 4. Birden fazla sÃ¼tuna gÃ¶re gruplama
print("4. BÄ°RDEN FAZLA SÃœTUNA GÃ–RE GRUPLAMA")
print("-" * 40)
coklu_grup = df.groupby(['BÃ¶lÃ¼m', 'YaÅŸ'])['Not'].mean()
print("BÃ¶lÃ¼m ve YaÅŸ'a gÃ¶re ortalama not:")
print(coklu_grup)
print()

# 5. Ã–zel fonksiyonlar
print("5. Ã–ZEL FONKSÄ°YONLAR")
print("-" * 40)
def not_araligi(x):
    return x.max() - x.min()

aralik = df.groupby('BÃ¶lÃ¼m')['Not'].apply(not_araligi)
print("BÃ¶lÃ¼mlere gÃ¶re not aralÄ±ÄŸÄ± (max - min):")
print(aralik)
````

## 1ï¸âƒ£8ï¸âƒ£ Pandas Merge ve Concat - Veri BirleÅŸtirme

Birden fazla DataFrame'i birleÅŸtirmek veri analizinde Ã§ok Ã¶nemlidir!


````python
# Pandas Merge ve Concat

print("=" * 60)
print("PANDAS MERGE VE CONCAT - VERÄ° BÄ°RLEÅTÄ°RME")
print("=" * 60)
print()

# Ã–rnek DataFrame'ler
df1 = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'Ä°sim': ['Ahmet', 'AyÅŸe', 'Mehmet', 'Zeynep'],
    'YaÅŸ': [20, 21, 19, 20]
})

df2 = pd.DataFrame({
    'ID': [1, 2, 3, 5],
    'Not': [85, 92, 78, 95],
    'BÃ¶lÃ¼m': ['Bilgisayar', 'Elektrik', 'Bilgisayar', 'EndÃ¼stri']
})

print("1. MERGE (BÄ°RLEÅTÄ°RME)")
print("-" * 40)
print("DataFrame 1:")
print(df1)
print("\nDataFrame 2:")
print(df2)

# Inner join (ortak ID'ler)
print("\nInner Join (ortak ID'ler):")
inner = pd.merge(df1, df2, on='ID', how='inner')
print(inner)

# Left join (sol DataFrame'in tÃ¼m satÄ±rlarÄ±)
print("\nLeft Join (sol DataFrame'in tÃ¼mÃ¼):")
left = pd.merge(df1, df2, on='ID', how='left')
print(left)

# Outer join (tÃ¼m satÄ±rlar)
print("\nOuter Join (tÃ¼m satÄ±rlar):")
outer = pd.merge(df1, df2, on='ID', how='outer')
print(outer)
print()

# 2. Concat (dikey birleÅŸtirme)
print("2. CONCAT (DÄ°KEY BÄ°RLEÅTÄ°RME)")
print("-" * 40)
df3 = pd.DataFrame({
    'ID': [5, 6],
    'Ä°sim': ['Can', 'Elif'],
    'YaÅŸ': [22, 19]
})
print("Yeni DataFrame:")
print(df3)
print("\nDikey birleÅŸtirme:")
birlesik = pd.concat([df1, df3], ignore_index=True)
print(birlesik)
````

````python
# PRATÄ°K PROJE 2: SatÄ±ÅŸ Verisi Analizi

print("=" * 60)
print("PROJE 2: SATIÅ VERÄ°SÄ° ANALÄ°ZÄ°")
print("=" * 60)
print()

# 1. SatÄ±ÅŸ verisi oluÅŸturma
print("1. SATIÅ VERÄ°SÄ° OLUÅTURMA")
print("-" * 40)
np.random.seed(42)

urunler = ['Laptop', 'Telefon', 'Tablet', 'KulaklÄ±k', 'Kamera']
kategoriler = ['Elektronik', 'Elektronik', 'Elektronik', 'Aksesuar', 'Elektronik']

satis_verisi = pd.DataFrame({
    'Tarih': pd.date_range('2024-01-01', periods=100, freq='D'),
    'ÃœrÃ¼n': np.random.choice(urunler, 100),
    'Kategori': np.random.choice(kategoriler, 100),
    'Miktar': np.random.randint(1, 10, 100),
    'Birim_Fiyat': np.random.randint(100, 2000, 100),
    'SatÄ±ÅŸ_Temsilcisi': np.random.choice(['Ali', 'AyÅŸe', 'Mehmet', 'Zeynep'], 100)
})

satis_verisi['Toplam_Tutar'] = satis_verisi['Miktar'] * satis_verisi['Birim_Fiyat']

print(f"Toplam satÄ±ÅŸ kaydÄ±: {len(satis_verisi)}")
print(f"\nÄ°lk 5 satÄ±ÅŸ:")
print(satis_verisi.head())
print()

# 2. Toplam satÄ±ÅŸ analizi
print("2. TOPLAM SATIÅ ANALÄ°ZÄ°")
print("-" * 40)
toplam_satis = satis_verisi['Toplam_Tutar'].sum()
ortalama_satis = satis_verisi['Toplam_Tutar'].mean()
print(f"Toplam satÄ±ÅŸ: {toplam_satis:,.2f} TL")
print(f"Ortalama satÄ±ÅŸ: {ortalama_satis:,.2f} TL")
print(f"En yÃ¼ksek satÄ±ÅŸ: {satis_verisi['Toplam_Tutar'].max():,.2f} TL")
print(f"En dÃ¼ÅŸÃ¼k satÄ±ÅŸ: {satis_verisi['Toplam_Tutar'].min():,.2f} TL")
print()

# 3. ÃœrÃ¼n bazlÄ± analiz
print("3. ÃœRÃœN BAZLI ANALÄ°Z")
print("-" * 40)
urun_analiz = satis_verisi.groupby('ÃœrÃ¼n').agg({
    'Toplam_Tutar': ['sum', 'mean', 'count']
}).round(2)
urun_analiz.columns = ['Toplam_SatÄ±ÅŸ', 'Ortalama_SatÄ±ÅŸ', 'SatÄ±ÅŸ_SayÄ±sÄ±']
print(urun_analiz.sort_values('Toplam_SatÄ±ÅŸ', ascending=False))
print()

# 4. Temsilci bazlÄ± analiz
print("4. TEMSÄ°LCÄ° BAZLI ANALÄ°Z")
print("-" * 40)
temsilci_analiz = satis_verisi.groupby('SatÄ±ÅŸ_Temsilcisi')['Toplam_Tutar'].sum().sort_values(ascending=False)
print("Temsilcilere gÃ¶re toplam satÄ±ÅŸ:")
print(temsilci_analiz)
print(f"\nEn baÅŸarÄ±lÄ± temsilci: {temsilci_analiz.idxmax()}")
print()

# 5. AylÄ±k satÄ±ÅŸ analizi
print("5. AYLIK SATIÅ ANALÄ°ZÄ°")
print("-" * 40)
satis_verisi['Ay'] = satis_verisi['Tarih'].dt.month
aylik_satis = satis_verisi.groupby('Ay')['Toplam_Tutar'].sum()
print("AylÄ±k satÄ±ÅŸ toplamlarÄ±:")
print(aylik_satis)
print(f"\nEn yÃ¼ksek satÄ±ÅŸ yapÄ±lan ay: {aylik_satis.idxmax()}. ay")
print()

print("âœ… Proje tamamlandÄ±! Bu tÃ¼r analizler iÅŸ dÃ¼nyasÄ±nda Ã§ok Ã¶nemlidir.")
````
