Şimdi daha anlamlı ve çok boyutlu bir veri oluşturup, methodları tekrar edelim. Böylece veriye daha fazla anlam yükleriz hem de diğer methodlar için çok boyutlu bir tabloya ihtiyacımız var. 

```python
#Sözlük kullanarak dataframe yapmaya çalışalım
sozluk1={"NAME":["AHMET", "MEHMET","SEVGİ","ATIL"], "SPORT":["RUNNİNG","SWİMİNG","RUNNİNG","BASKETBALL"],
         "CALORİ":[100,200,300,400]
        }
```

```python
#İndex default olarak 0 dan başlayacaktır ama ben indexi 1den başlatmak istedim. 
#Siz isterseniz her satıra farklı isimlendirebilirsiniz.
df1 = pd.DataFrame(sozluk1, index=[1,2,3,4])
```

```python
df1
```

```python
sozluk2={"NAME":["SEVGİ", "MEHMET","HAKAN","METE"], "SPORT":["JOGGİNG","CLİMBİNG","RUNNİNG","FOOTBALL"],
         "CALORİ":[100,350,400,500]
        }
```

```python
df2 = pd.DataFrame(sozluk2, index= [5,6,7,8])
```

```python
df2
```

```python
sozluk3={"NAME":["CANER", "SEVGİ","ÖMÜR","ATIL"], "SPORT":["RUNNİNG","SOCCER","RUNNİNG","YOGA"],
         "CALORİ":[500,450,500,700]
        }
df3=pd.DataFrame(sozluk3, index=[9,10,11,12])
df3
```

Concatenation: birleştirme işlemi

```python
#pd.concat([df1,df3,df2])   Verilerin sırası birleştirmenin hangi sıraya göre yapılacağına karar verir.
pd.concat([df1,df2,df3])
```

```python
#Merge 2 tabloyu belirli bir sütun üzerine birleştirmeyi hedefler.
#Elimizde 3 tablo olduğu için df1 ve df2 yi belirli NAME sütunu üzerinde birleştireceğiz.
df12=pd.merge(df1, df2, on="NAME")
df12
```

```python
#df1 ve df2 yi birleştirerel df12 yi elde ettik. Ardından df3 ile birleştirdik.
df123=pd.merge(df12,df3, on ="NAME")
df123
```

Merge kullanarak 3 data frame in birleşimini elde edelim. Yeni şeyler deneyelim 

```python
df12=pd.merge(df1,df2, on="SPORT", how ='left')
df12
```

```python
df_new=pd.merge(df12,df3, on="SPORT", how='right')
df_new
```

```python
#Veri setinde eksik veri olup olmadığını gösterir. Eksik veriyse(NaN) true döndürür
df_new.isnull()
```

```python
#Her bir sütunda kaç tane eksik veri olduğunu gösterir.
df_new.isnull().sum()
```

```python
#Oluşturduğumuz veri setinin bilmemiz gereken tüm istatiksel verileri yazdırır
df_new.describe()
```

```python
#Veri setinde NaN sayısı ve sütunların tuttuğu veri tiplerini gösterir
df_new.info()
```

```python
df_new
```

```python
#Tüm nan veriler önümüzde. Eksik verilerden kurtulalım önce. dropna methoduyla eksik verinin olduğu tüm satırlar ortadan kalkar.
df=df_new.dropna(inplace=False)
```
